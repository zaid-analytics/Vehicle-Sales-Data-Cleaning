```python
sales_data["odometer"] = sales_data["odometer"].fillna(68320.0)
```

```python
# Check column data type before filling null values
print(sales_data["saledate"].dtype)
```

```python
# fill null values
# print(sales_data["model"])
sales_data["make"] = sales_data["make"].fillna(sales_data["make"].mode()[0])
print(sales_data["make"].isnull().sum())
```

```python
sales_data["model"] = sales_data["model"].fillna(sales_data["model"].mode()[0])
print(sales_data["model"].isnull().sum())
```

```python
sales_data["trim"] = sales_data["trim"].fillna(sales_data["trim"].mode()[0])
print(sales_data["trim"].isnull().sum())
```

```python
sales_data["body"] = sales_data["body"].fillna(sales_data["body"].mode()[0])
print(sales_data["body"].isnull().sum())
```

```python
sales_data["transmission"] = sales_data["transmission"].fillna(sales_data["transmission"].mode()[0])
print(sales_data["transmission"].isnull().sum())
```

```python
sales_data["vin"] = sales_data["vin"].fillna(sales_data["vin"].mode()[0])
print(sales_data["vin"].isnull().sum())
```

```python
# we will find mean value if we have int/float column
print(sales_data["condition"].mean().round(1))
sales_data["condition"].fillna(30.7, inplace = True)
 # in advance way
sales_data["condition"].fillna(sales_data["condition"].mean(), inplace = True)
```

```python
sales_data["color"] = sales_data["color"].fillna(sales_data["color"].mode()[0])
print(sales_data["color"].isnull().sum())
```

```python
sales_data["interior"] = sales_data["interior"].fillna(sales_data["interior"].mode()[0])
print(sales_data["interior"].isnull().sum())
```

```python
sales_data["mmr"] = sales_data["mmr"].fillna((sales_data["mmr"]).mean())
print(sales_data["mmr"].isnull().sum())
print(sales_data["mmr"].mean())         # mean value will same after filling the null values
```

```python
sales_data["sellingprice"] = sales_data["sellingprice"].fillna((sales_data["sellingprice"]).mean())
print(sales_data["sellingprice"].isnull().sum())
print(sales_data["sellingprice"].mean())  
```

```python
sales_data["saledate"] = pd.to_datetime(sales_data["saledate"], errors = "coerce", utc = True)  #change data type 
print(sales_data["saledate"].dtype)
sales_data["saledate"] = sales_data["saledate"].ffill()
```

```python
after_fill = sales_data.isnull().sum()
print(after_fill)
print(" Eventually Data has cleaned")
```

```python
# save file in current folder

sales_data.to_csv("clean_csv.csv", index = False)
```
